```python
import numpy as np

# Define Q1 (25th percentile) and Q3 (75th percentile)
Q1 = df_new["price"].quantile(0.01)
Q3 = df_new["price"].quantile(0.99)

# Compute IQR
IQR = Q3 - Q1

# Define lower and upper bound
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Find outliers
outliers = df_new[(df_new["price"] < lower_bound) | (df_new["price"] > upper_bound)]
df_new
```

```python
df_new.isnull().sum()
```

```python
df_no_outliers = df_new[(df_new["price"] >= lower_bound) & (df_new["price"] <= upper_bound)]

# Display the filtered DataFrame
df_no_outliers
```

```python
df_no_outliers["availability_365"]==0
```

```python
# Filter rows where 'availability_365' is 0
availability_zero_rows = df_no_outliers[df_no_outliers["availability_365"] == 0]

# Display the rows
availability_zero_rows
```

```python
# Filter rows where 'availability_365' is not 0
df_available_365 = df_no_outliers[df_no_outliers["availability_365"] != 0]

# Display the new DataFrame
df_available_365
```

```python
plt.figure(figsize=(10, 6))

# Plot histogram for Price
plt.hist(df_available_365["price"], bins=30, color='blue', alpha=0.7)
plt.title("Price Distribution of Airbnb Listings")
plt.xlabel("Price (in $)")
plt.ylabel("Frequency")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Plotting a box plot for Room Type vs Price
plt.figure(figsize=(10, 6))
sns.boxplot(x='room_type', y='price', data=df_available_365, palette='Set2')

# Adding title and labels
plt.title('Room Type vs Price')
plt.xlabel('Room Type')
plt.ylabel('Price (in $)')
plt.grid(True)

# Show plot
plt.show()
```

```python
# Calculate the average price for each room type
avg_price_by_room_type = df_available_365.groupby('room_type')['price'].mean().reset_index()

# Plotting a bar plot for Room Type vs Average Price
plt.figure(figsize=(10, 6))
sns.barplot(x='room_type', y='price', data=avg_price_by_room_type, palette='Set2')

# Adding title and labels
plt.title('Average Price by Room Type')
plt.xlabel('Room Type')
plt.ylabel('Average Price (in $)')
plt.grid(True)

# Show plot
plt.show()
```

```python
# Calculate average price and number of listings per neighborhood group
neighbourhood_stats = df_available_365.groupby('neighbourhood_group').agg(
    avg_price=('price', 'mean'),
    num_listings=('price', 'size')
).reset_index()

# Plotting the scatter plot
plt.figure(figsize=(12, 6))
sns.scatterplot(data=neighbourhood_stats, x='neighbourhood_group', y='avg_price', 
                size='num_listings', hue='num_listings', palette='coolwarm', sizes=(50, 500), legend=False)

# Adding title and labels
plt.title('Average Price, Neighborhood Group, and Number of Listings')
plt.xlabel('Neighborhood Group')
plt.ylabel('Average Price (in $)')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.grid(True)

# Show plot
plt.show()
```

```python
# Convert 'last_review' to datetime format
df_new['last_review'] = pd.to_datetime(df_new['last_review'], errors='coerce')

# Extract month and year
df_new['review_month'] = df_new['last_review'].dt.month
df_new['review_year'] = df_new['last_review'].dt.year

# Display the first few rows to verify the changes
print(df_new[['last_review', 'review_month', 'review_year']].head())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Group by 'review_month' and count the number of reviews
month_reviews = df_new.groupby('review_month').size().reset_index(name='review_count')

# Plot the bar chart
plt.figure(figsize=(10, 6))
sns.barplot(data=month_reviews, x='review_month', y='review_count', palette='viridis')

# Set labels and title
plt.title('Number of Reviews by Month', fontsize=16)
plt.xlabel('Month', fontsize=12)
plt.ylabel('Number of Reviews', fontsize=12)
plt.xticks(ticks=range(12), labels=['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])

# Show the plot
plt.tight_layout()
plt.show()
```
